```python
from transformers import pipeline
```

## **As the name suggests, sentiment analysis would look for the emotions carried in the argument passed in it.**

```python
classifier = pipeline('sentiment-analysis')
classifier('I feel good today')
```

```python
classifier = pipeline('sentiment-analysis')
classifier('I dont feel so good')
```

```python
classifier = pipeline('sentiment-analysis')
classifier('Everything seems confusing')
```

```python
classifier = pipeline('sentiment-analysis')
classifier(['I am happy!',
            'I am sad!'])
```

# **As the name suggests, One Shot Classification - evaluates the statement against the labels passed in the arguments.**

```python
classifier = pipeline('zero-shot-classification')
classifier('I feel good', candidate_labels = ['emotions','education','construction'])
```

```python
classifier = pipeline('zero-shot-classification')
classifier('This is a course on Transformers Library', candidate_labels = ['emotions','education','construction'])
```

```python
classifier = pipeline('text-generation')
classifier('In this course, we will learn to')
```

# **We would be now using a custom model to do our tasks. If we do not specify the model, default is OPENAI GPT 2.0**

```python
from transformers import pipeline

classifier = pipeline('text-generation', model='distilgpt2')
classifier('This course is designed for an asipiring',
           max_length = 100,
           num_return_sequences = 2)
```

# **If we want to implement something with fill in the blanks or questionnarie use cases. The fill-mask technique helps. We can train the model based on domain specific articles and it would fill the masks with the right value**

```python
from transformers import pipeline

unmasked_data = pipeline('fill-mask')
unmasked_data('Paris is the <mask> of France')
```

```python
from transformers import pipeline

unmasked_data = pipeline('fill-mask')
unmasked_data('Paris is the <mask> of France',top_k=3)
```
